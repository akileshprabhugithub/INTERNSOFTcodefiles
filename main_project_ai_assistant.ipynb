```python
engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
engine.setProperty("voice",voices[1].id)
```

```python
def speak(text):
    engine.say(text)
    engine.runAndWait()
```

```python
def wishMe():
    hour = datetime.datetime.now().hour
    if hour >=0 and  hour<12:
        speak("hello, good morning Mr. Akilesh")
        print("hello, good morning Mr. Akilesh")
    elif hour >= 12 and hour <4:
        speak("hello, good afternoon Mr. Akilesh")
        print("hello, good afternoon Mr. Akilesh")
    elif hour >= 4 and hour <19:
        speak("hello, good evening Mr. Akilesh")
        print("hello, good evening Mr. Akilesh")
    else:
        speak("its about time you had some sleep")
        print("its about time you had some sleep")
        
        
        
```

```python
def takeCommand():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        r.adjust_for_ambient_noise(source)
        print("i am listening")
        audio = r.listen(source)
        
        try:
            statement = r.recognize_google(audio,language='en-in')
            print(f"user said : {statement}\n")
            
        except Exception as e:
            speak("pardon me, please repeat that again")
            return "None"
        return statement 
        
```
